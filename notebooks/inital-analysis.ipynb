```python
# LLM Generated code

import pandas as pd, re

abbrs = {
    'F. P.', 'W', 'P', 'F. B.', 'S. P.', 'S. Ps.', 'G. P',
    'B. S.', 'M. S.', 'M. P', 'B. R.', 'B. P.', 'L. B', 'P. O.'
}

pattern = re.compile(r'^[A-Z](?:\.\s?)?(?:[A-Z]\.?)?$')  # one- or two-letter codes

osdf = gb1900
df_clean = osdf[~(osdf['final_text'].isin(abbrs) | osdf['final_text'].str.match(pattern))]
df_clean.to_csv('os_1880_no_abbr.csv', index=False)
```

```python
#LLM Generated Code

# 1. compute the value counts
counts = noabbr['final_text'].value_counts()

# 2. cumulative coverage
cum_pct = counts.cumsum() / counts.sum()

# 3. pick threshold and find how many labels we need
threshold = 0.90
num_labels = (cum_pct <= threshold).sum()
print(f"{num_labels} labels cover {threshold*100:.0f}% of all records.")

# 4. slice out just those top labels
top_counts = counts.iloc[:num_labels]

# 5. build a DataFrame and export
df_top = top_counts.rename_axis('final_text').reset_index(name='count')
df_top.to_csv('top_final_text_90pct.csv', index=False)

print("Exported → top_final_text_90pct.csv")
```
