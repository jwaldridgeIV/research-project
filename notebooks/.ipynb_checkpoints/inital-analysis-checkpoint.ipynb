```python
oil = gb1900[gb1900['final_text'].str.contains('Oil')]
oil.sample(5)
```

```python
import folium
```

```python
uk_map = folium.Map(location=[53, -1], zoom_start=6)
uk_map
```

```python
def create_map_markers(row, map_name):
    folium.Marker(location=[row['latitude'], row['longitude']], popup=row['final_text']).add_to(map_name)
```

```python
empty_map = folium.Map(location=[53, -1], zoom_start=6)
```

```python
oil.apply(
    create_map_markers, 
    map_name=empty_map, 
    axis='columns' 
)

empty_map
```
